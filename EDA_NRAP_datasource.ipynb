Overall, these projects have played a vital role in improving rural access to services, enhancing community resilience, and promoting sustainable development in Afghanistan.

## 6.0 Central Tendency

Central tendency in statistics refers to the measure that represents the central or typical value of a dataset. It is also known as the average or the middle value of a set of numbers. There are three common measures of central tendency:

***Mean:*** The mean is the sum of all the values in the dataset divided by the total number of values.

***Median:*** The median is the middle value of the dataset when the values are arranged in order from lowest to highest (or vice versa).

***Mode:*** The mode is the value that appears most frequently in the dataset.

> **Note:** Each of these measures has its own strengths and weaknesses, and which one is used depends on the nature of the data and the purpose of the analysis.

## 6.1 Skewness
Skew in statistics, refers to the measure of the asymmetry of a probability distribution around its mean.

A symmetrical distribution has the same shape on both sides of the mean, while a skewed distribution has a longer tail on one side compared to the other.

There are two types of skewness:

***Positive Skewness:*** Also known as right-skewed, it occurs when the tail of the distribution extends more towards the right side. In a positively skewed distribution, the mean is greater than the median.

***Negative Skewness:*** Also known as left-skewed, it occurs when the tail of the distribution extends more towards the left side. In a negatively skewed distribution, the mean is less than the median.

> **Note:** Skewness is an important measure in statistical analysis because it affects the interpretation of data and can influence the choice of statistical methods used for analysis.

```python
numeric_cols = data.select_dtypes(include='float').columns
skewness = data[numeric_cols].skew()
skewness
```

> Note: A value less than -1 is skewed to the left;  greater than 1 is skewed to the right.
A value between -1 and 1 is symmetric.

# 8 Correlation
Correlation refers to the relationship between two or more variables. Correlation is a measure of the degree to which two or more variables are related to each other.

There are two main types of correlation:

***Positive correlation:*** In positive correlation, as one variable increases, the other variable also tends to increase. This means that the two variables move in the same direction.

***Negative correlation:*** In negative correlation, as one variable increases, the other variable tends to decrease. This means that the two variables move in opposite directions.


```python
# generate a correlation  for data dataframe
round(data.corr(),2)
```

```python
# generate heatmap to see the correlation
plt.figure(figsize=(14,8)) # Set the figure size to be 14 inches wide and 8 inches tall

# Create a heatmap of the correlation matrix for the data, with annotations and font size 11
sns.heatmap(data.corr(), annot=True, annot_kws={"size": 11})

sns.set_style('white') # Set the style of the plot to 'white'

plt.xticks(fontsize=10) # Set the font size for the x-axis tick labels to 10

plt.yticks(fontsize=10) # Set the font size for the y-axis tick labels to 10

plt.show() # Display the plot
```

> **Note:** A correlation coefficient of **-1** indicates a perfect negative correlation, a correlation coefficient of **0** indicates no correlation, and a correlation coefficient of **1** indicates a perfect positive correlation.

> A ***weak correlation*** is one in which the correlation coefficient is close to zero. This means that there is little or no relationship between the two variables. For example, if we were to measure the correlation between a person's height and their favorite color, we would likely find a weak correlation, as these two variables are likely unrelated.

> A ***moderate correlation*** is one in which the correlation coefficient is somewhere between -1 and 1, but closer to zero than to either extreme. This means that there is a moderate degree of relationship between the two variables. # 9.0 Time Series Analysis
Time series analysis is a statistical method used to analyze data that varies over time. It is often used to analyze and predict patterns in financial, economic, social, or environmental data.

The basic premise of time series analysis is to use historical data to identify patterns and trends, and then use that information to make predictions about future data points. Time series analysis can involve a wide variety of statistical techniques, including linear regression, autoregression, moving average models, exponential smoothing, and more.

Some common applications of time series analysis include:

    - Stock market forecasting
    - Weather forecasting
    - Economic forecasting
    - Traffic forecasting
    - Energy demand forecasting
Overall, time series analysis can be a powerful tool for understanding trends and making predictions about future events based on past data.


## 9.1 Create DataFrame
In this section, we create a new dataframe for time series analysis to capture a single column of **date** along with five numerical columns. The data is then aggregated, resulting in a single record for each year across all five columns.

```python
# create a time_data dataframe to only capture aggregated data per completion month for bellow filve
# columns
time = data[['ActualCompletionDate', 'Road(KM)', 'Bridge(M)', 'Cost(USD)', 'LabourDay','Structure(M)']].copy()

#time_data.loc[:, 'CompletionYear'] = pd.to_datetime(time_data['ActualCompletionDate']).dt.year
time.loc[:, 'CompletionYear'] = pd.to_datetime(time['ActualCompletionDate'], format='%Y')
time = time.drop(columns='ActualCompletionDate')
time.set_index('CompletionYear', inplace=True)
time_data = time.resample('M').sum() # sum the data per month
time_data
```

## 9.2 Visualize Time Series Data

```python
# Apply seasonal decomposition to each variable in the DataFrame
decompositions = {}
for column in time_data.columns:
    decomposition = seasonal_decompose(time_data[column], model='additive')
    decompositions[column] = decomposition

# Plot the decomposition results for each variable
fig, axes = plt.subplots(nrows=len(time_data.columns), ncols=1, figsize=(10, 20), sharex=True)
for i, column in enumerate(time_data.columns):
    axes[i].plot(time_data.index, time_data[column], label=f'Original {column}', color='blue')
    axes[i].legend(loc='best')
    axes[i].set_title(f'{column} Time Series Decomposition', fontsize=14)

plt.tight_layout()
plt.show()
```

>Note: The output of the above code will be a time_data dataframe that captures aggregated data per completion month for five columns - Road(KM), Bridge(M), Cost(USD), LabourDay, and Structure(M). The dataframe is then resampled at a monthly frequency and the seasonal decomposition is applied to each variable in the dataframe. The seasonal decomposition results for each variable are then plotted in separate subplots, with the original time series and its decomposed components shown. The resulting plots provide insights into the underlying patterns and trends in each variable over time.

```python
# Apply seasonal decomposition to each variable in the DataFrame
decompositions = {}
for column in time_data.columns:
    decomposition = seasonal_decompose(time_data[column], model='additive')
    decompositions[column] = decomposition

# Extract the trend, seasonal, and residual components for each variable
trends = {}
seasonals = {}
residuals = {}
for column in time_data.columns:
    decomposition = decompositions[column]
    trends[column] = decomposition.trend
    seasonals[column] = decomposition.seasonal
    residuals[column] = decomposition.resid

# Plot the decomposition results for each variable
fig, axes = plt.subplots(nrows=len(time_data.columns), ncols=1, figsize=(10, 20), sharex=True)
for i, column in enumerate(time_data.columns):
    axes[i].plot(time_data.index, time_data[column], label='Original', color='blue')
    axes[i].plot(trends[column].index, trends[column], label='Trend', color='orange', linewidth=2)
    axes[i].plot(seasonals[column].index, seasonals[column], label='Seasonal', color='green', linewidth=2)
    axes[i].plot(residuals[column].index, residuals[column], label='Residual', color='red', linewidth=2)
    axes[i].legend(loc='best')
    axes[i].set_title(f'{column} Time Series Decomposition', fontsize=14)

plt.tight_layout()
plt.show()
```

> NoteThe output of the above code will be a series of plots that show the decomposition of each variable in the time_data dataframe into its trend, seasonal, and residual components. For each variable, the original time series is plotted in blue, while the trend, seasonal, and residual components are plotted in orange, green, and red, respectively. The plots provide insights into how each variable's trend, seasonal, and residual components contribute to the overall behavior of the time series.

```python
# Resample the data to yearly frequency
time_data = time.resample('Y').sum()

# Apply seasonal decomposition to RoadKM, BridgeM,Cost(USD), LabourDay
decomposition1 = seasonal_decompose(time_data['Road(KM)'], model='additive')
decomposition2 = seasonal_decompose(time_data['Bridge(M)'], model='additive')
decomposition3 = seasonal_decompose(time_data['Cost(USD)'], model='additive')
decomposition4 = seasonal_decompose(time_data['LabourDay'], model='additive')
decomposition5 = seasonal_decompose(time_data['Structure(M)'], model='additive')
# Extract the trend, seasonal, and residual components for RoadKM,BridgeM
trend1 = decomposition1.trend.cumsum()
trend2 = decomposition2.trend.cumsum()
trend3 = decomposition3.trend.cumsum()
trend4 = decomposition4.trend.cumsum() 
trend5 = decomposition5.trend.cumsum() 
fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(10, 16), sharex=True)

axes[0].plot(trend1.index, trend1, label='Road(KM)', color='blue', linewidth=2)
axes[0].legend(loc='best')
axes[0].set_title('Cumulative Sum of Road(KM) Time Series', fontsize=14)
axes[0].set_ylabel('Cumulative Sum of Road (KM)', fontsize=12)
axes[0].set_xticks(trend1.index)
axes[0].set_xticklabels(trend1.index.year, rotation=45)

axes[1].plot(trend2.index, trend2, label='Bridge(M)', color='green', linewidth=2)
axes[1].legend(loc='best')
axes[1].set_title('Cumulative Sum of Bridge(M) Time Series', fontsize=14)
axes[1].set_ylabel('Cumulative Sum of Bridge (M)', fontsize=12)
axes[1].set_xticks(trend2.index)
axes[1].set_xticklabels(trend2.index.year, rotation=45)

axes[2].plot(trend3.index, trend3, label='Cost(USD)', color='red', linewidth=2)
axes[2].legend(loc='best')
axes[2].set_title('Cumulative Sum of Cost(USD) Time Series', fontsize=14)
axes[2].set_ylabel('Cumulative Sum of Cost In $100 Million', fontsize=12)
axes[2].set_xticks(trend3.index)
axes[2].set_xticklabels(trend3.index.year, rotation=45)

axes[3].plot(trend4.index, trend4, label='LabourDay', color='yellow', linewidth=2)
axes[3].legend(loc='best')
axes[3].set_title('Cumulative Sum of LabourDay Time Series', fontsize=14)
axes[3].set_ylabel('Cumulative Sum of LabourDay (Million)', fontsize=12)
axes[3].set_xticks(trend4.index)
axes[3].set_xticklabels(trend4.index.year, rotation=45)

axes[4].plot(trend5.index, trend5, label='Structure(M)', color='yellow', linewidth=2)
axes[4].legend(loc='best')
axes[4].set_title('Cumulative Sum of Structure Time Series', fontsize=14)
axes[4].set_ylabel('Cumulative Sum of Structure (Meter)', fontsize=12)
axes[4].set_xticks(trend5.index)
axes[4].set_xticklabels(trend5.index.year, rotation=45)

plt.tight_layout()
plt.show()
```

> Note: The output of this code will be a plot with five subplots, each showing the cumulative sum of one of the variables in the time_data DataFrame (Road(KM), Bridge(M), Cost(USD), LabourDay, and Structure(M)).

>For each variable, the code first applies seasonal decomposition using an additive model. It then extracts the trend component and takes the cumulative sum of the trend to obtain the cumulative sum of the variable.

>The resulting plot shows the cumulative sum of each variable over time, with each year represented by a single point on the x-axis. The y-axis shows the cumulative sum of the variable. Each subplot has a title indicating the variable being plotted and a legend indicating the color used for that variable. The x-axis labels show the years being plotted.
