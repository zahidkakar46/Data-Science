# 1 Removing Unnecessary Columns

```python
#create function
def remove_columns(df, columns):
    """
    Remove specified columns from a pandas dataframe if they exist.

    Parameters:
    df (pandas dataframe): The dataframe to remove columns from.
    columns (list): A list of column names to remove.

    Returns:
    The modified dataframe with the specified columns removed, or the original dataframe if the specified columns do not exist.
    """
    columns_to_remove = [col for col in columns if col in df.columns]
    if columns_to_remove:
        df = df.drop(columns_to_remove, axis=1)
    return df
```

```python
# call the function
unnecessary_columns = ['ProvinceID', 'DistrictID', 'StatusID', 'ContractTypeID', 'ProjectID', 'ContractorID']
df = remove_columns(df, unnecessary_columns)
df.head(5)
```
